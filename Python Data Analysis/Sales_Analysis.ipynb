```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Diwali sales Data.csv', encoding='unicode_escape')
```

```python
df.head()
```

# Data Cleaning

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df=df.astype({"User_ID":"int32",
             "Marital_Status":"int8",
             "Age":"int8"
            })
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df[['Age','Orders','Amount']].describe()
```

# Exploratory Data Analysis




## Gender

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df.groupby(["Gender"]).agg({"Orders":"sum"}).plot.bar(ylabel="Orders",figsize=(5,5))
```

```python
df.groupby(["Age Group","Gender"]).agg({'Orders':'sum'}).plot.bar(color=['Red'])
```

**Therefore from the above graph we can see that most of the buyers are females.**

## Age

```python
df.groupby(["Age Group"]).agg({'Orders':'sum'}).sort_values(by="Orders").plot.bar(ylabel="Orders")
```

**From the above graph we can see that highest number of orders are placed by the age group 26yrs-35yrs.**

## State

```python
df.groupby(["State"]).agg({'Orders':'sum'}).sort_values(by="Orders").plot.bar(ylabel="Orders")
```

```python
df.groupby(["State"]).agg({'Amount':'sum'}).sort_values(by="Amount").plot.bar(ylabel="Amount",color="#59c977")
```

**For the above graphs we can conclude that highest number of orders were from UP,Maharashtra and Karnataka.**

## Marital Status

```python
df['Marital_Status'] = df['Marital_Status'].replace([1, 0], ['Unmarried', 'Married'])
```

```python
df.head()
```

```python
sns.countplot(data=df,x="Marital_Status")
```

```python
sales=df.groupby(["Marital_Status","Gender"],as_index=False).agg({'Amount':'sum'})
sns.barplot(data =sales,x ="Marital_Status", y ="Amount",hue ="Gender")
```

**From the above graphs we can conclude that most of the buyers are married (women).**

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.countplot(data=df,x="Occupation")
```

**The above graph tells us that most of the buyers are from IT Sectors,Healthcare and Aviation.**

## Product Category

```python
sns.set(rc={'figure.figsize':(27,5)})
sns.countplot(data=df,x="Product_Category")
```

**From the above graph we can see that most of the items purchased are from Clothing,food and electronics.**

## Conclusion:

**Married woman age 26-35 yrs from UP, Maharashtra and Karnataka working in IT,Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics.**
